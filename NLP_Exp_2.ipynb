#Web Scrapping

```python
import requests
from bs4 import BeautifulSoup

#Make a request to the website      
page = requests.get("https://jamesclear.com/best-books/best-selling")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
display = soup.find_all("strong")
#Find all the elements on the page with a specific html tag

for d in display:
  print(d.text)
```

```python
# Find all the elements on the page that have a specific class
soup.find_all(class_="footnote")
```
