```python
import pandas as pd
import nltk
import spacy
```

```python
df = pd.read_csv('spam.csv')
df = df.text
df
```

```python
# Punctuational Removal
import string
df_punc = df.str.translate(str.maketrans("","", string.punctuation))
df_punc
```

```python
# Lowering the text
df_lower = df_punc.str.lower()
df_lower
```

```python
import re
df_digit = re.sub(r'\d+','',str(df_lower))
print(df_digit)
```

```python
#Stopwords download
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words_nltk = set(stopwords.words('english'))
print(stop_words_nltk)
```

```python
#removing stopwords
from gensim.parsing.preprocessing import remove_stopwords
df_without_stop = remove_stopwords(df_digit)
print(df_without_stop)
```

```python
# Tokenization
from nltk.tokenize import word_tokenize
nltk.download('punkt')
tokenized_df = word_tokenize(df_without_stop)
print(tokenized_df)
```

```python
#Stemming
from pprint import pprint
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
stemmer= PorterStemmer()
print("Before Stemming")
print(df_without_stop)
print("After Stemming")
for word in tokenized_df:
  print(stemmer.stem(word),end=" ")
```

```python
# lemetizing the text
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
print("After Lemetization")
lemm = WordNetLemmatizer()
for word in tokenized_df:
  print(lemm.lemmatize(word),end=" ")
```

```python
# POS tagging
nltk.download("averaged_perceptron_tagger")
print("POS Tagging using NLTK:")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

# get tf-df values
result = tfidf.fit_transform(df)

# get idf values
print('\nidf values:')
for e1, e2 in zip(tfidf.get_feature_names(), tfidf.idf_):
  print(e1, '=', e2)

# get indexing
print('\nWord indexes:')
print(tfidf.vocabulary_)

# display tf-idf values
print('\ntf-idf value:')
print(result)
```
